```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pylab as plt
import scipy.stats as st
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Intelligent-Systems-Phystech/psad/master/labs/lab1/data/sick.csv")
data
```

Сначала посмотрим, насколько данные нормальны. Теория говорит, что критерий Стьюдента значимости признака можно применять, если имеем дело с нормальной моделью. То есть, если $(y - \hat{y})\ |\ X \sim N(0, \sigma^2)$.

```python
model = sm.GLM.from_formula('Y~x1 + x2 + x3 + x4 + x5 + x6 + x7 + x8 + x9 + x10', 
                             data=data).fit()
```

```python
plt.hist(model.predict(data) - data['Y'], bins=30)
```

```python
print(st.shapiro(model.predict(data) - data['Y']))
_ = st.probplot(model.predict(data) - data['Y'], plot=plt)
plt.show()
```

Предположение о нормальности ошибок выполняется с большой натяжкой. Но других инструментов мы не знаем, поэтому воспользуемся критерием Стьюдента, а потом посмотрим, к чему привёл наш отбор признаков, улучшил ли он ситуацию.

Посмотрим, насколько значимых каждый признак по отдельности с помощью критерия Стьюдента.

**Критерий Стьюдента**

$$H_0:\ \ \beta_j = 0$$

$$H_1:\ \ \beta_j \neq 0$$

$$T = \frac{\hat{\beta_j}}{\sqrt{\frac{RSS}{n-k-1} (X^TX)^{-1}_{jj}}}$$

Нулевое распределение: $St(n-k-1)$.

p-value:

$$p = 1 - 2 F_{St(n-k-1)}(T)$$

$F_{St(n-k-1)}$ -- функция распределения для $St(n-k-1)$. 

В этой формуле мы $F_{St(n-k-1)}$ Домножаем на 2, так как отсекаем хвосты распределения Стьюдента с обеих сторон. 

```python
model.summary2()
```

Оценим полученные результаты: как видно из столбца $P>|z|$, все признаки кроме $x_1, x_3$ являются незначимыми. Удалим их.

```python
model = sm.GLM.from_formula('Y ~ x1 + x3',
                             data=data).fit()
model.summary2()
```

Log-Likelihood (который нам надо максимизировать) уменьшился несильно. Значит, можно сделать вывод о том, что удаление признаков почти не повлияло на качество.

Попробуем возвращать по одному удалённые признаки:

```python
formula = 'Y ~ x1 + x3 +'
for p in ['x2','x4','x5','x6','x7','x8', 'x9', 'x10']:
    model = sm.GLM.from_formula(formula+p, data=data).fit()
    #print (p, model.bic)
    print(model.summary2())
```

Видим, что если добавить любой из оставшихся признаков к $x_1, x_3$, то p-value покажет, что он избыточен при уровне значимости $\alpha = 0.05$. Также Log-Likelihood не увеличится.

Посмотрим теперь на линейную регрессию только с одним из признаков $x_1, x_3$.

```python
model = sm.GLM.from_formula('Y ~ x1',
                             data=data).fit()
model.summary2()
```

```python
model = sm.GLM.from_formula('Y ~ x3',
                             data=data).fit()
model.summary2()
```

В обоих случаях получились сильно хуже, чем в модели $Y \sim x_1 + x_3$.

Итак, считаем, что из 10 признаков значимы только $x_1, x_3$.
