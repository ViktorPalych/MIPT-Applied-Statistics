# Решение

Итак, мы имеем 2 выборки по 14 элементов: $X_1^{14} = (X_{1,1}, \dots, X_{1,14}), X_2^{14} = (X_{2,1}, \dots, X_{2,14}) $. $X_{1,i}, X_{2,i}$ -- времена восстановления температуры i-того пациента для стандартного одеала и экспериметнального соответственно. Пациент один и тот же для $X_{1,i}, X_{2,i}$, поэтому выборки связаны.

Чтобы понять, отличаются ли одеяла, можно проверить гипотезу о равенстве матожиданий (или медиан) случайных величин $X_1, X_2$. Нам ничего не известно о распределении времени восстановления, поэтому надо использовать более общие критерии. Из тех критериев, которые были на лекциях, лучше подходят Критерий знаковых рангов Уилкоксона для связанных выборок (медиана) и Двухвыборочный перестановочный критерий для связанных выборок (среднее).

### Двухвыборочный перестановочный критерий для связанных выборок

$$H_0:\ \ \mathbb{E} X_1=  \mathbb{E}  X_2$$

$$H_1:\ \ \mathbb{E} X_1 \neq \mathbb{E}  X_2$$

$$\text{Статистика}: D_i = X_{1i} - X_{2i},\ \  T(X_1^{14}, X_2^{14}) = \sum_{i=1}^{14}D_i$$

Нулевое распределение порождается перебором $2^n$ знаков перед $D_i$.

Распределение симметрично относительно нуля. При этом, критическая область -- это два одинаковых куска: края слева и справа. Поэтому, достигаемый уровень значимости: $1 - 2F_T\left( \big| \sum_{i=1}^{14}D_i \big| \right) $ ($F_T$ -- функция распределения статистики).

### Критерий знаковых рангов Уилкоксона для связанных выборок 

$$H_0:\ \ \text{med} X_1=  \text{med} X_2$$

$$H_1:\ \ \text{med} X_1 \neq \text{med} X_2$$

$$\text{Статистика}:\ \  W(X_1^{14}, X_2^{14}) = \sum_{i=1}^{14} \text{rank}(|X_{1i} - X_{2i} |)\cdot \text{sign}(X_{1i} - X_{2i})$$

Нулевое распределение табличное.

Аналогично, достигаемый уровень значимости: $1 - 2F_W\left( \big| \sum_{i=1}^{14} \text{rank}(|X_{1i} - X_{2i} |)\cdot \text{sign}(X_{1i} - X_{2i}) \big| \right) $ ($F_W$ -- функция распределения статистики).

Если же дополнительно известно, что распределение времени восстановления температуры можно аппроксимировать нормальным, то можно использовать критерий Стьюдента для связанных выборок (дисперсия нам вряд ли будет известна, поэтому Z-критерий не пойдёт). Он не так сильно чувствителен к отклонениям от нормальности, поэтому "нормальность" наших распределений можно оценить, построив qq-plot. Если же точки на нём не сильно отклоняются от прямой, то используем.

### Критерий Стьюдента для связанных выборок

$$H_0:\ \ \mathbb{E}  X_1=  \mathbb{E} X_2$$

$$H_1:\ \ \mathbb{E}  X_1 \neq \mathbb{E} X_2$$

$$\text{Статистика}:\ \  T(X_1^{14}, X_2^{14}) = \frac{(\overline{X_1} - \overline{X_2})\sqrt{n}}{S}$$

$$S = \sqrt{\frac{1}{n-1} \sum_{i=1}^{14} (D_i - \bar{D})^2},\ \  D_i = X_{1i} - X_{2i},\ \  \bar{D} = \sum D_i $$

Нулевое распределение здесь $St(n-1) = St(13)$. Достигаемый уровень значимости: $1 - 2F_{St(13)}\left(| T(X_1^{14}, X_2^{14}) |\right) $

Если не отклоняем гипотезу $H_0$, то это означает, что не отклоняем гипотезу о том, что между одеялами разницы нет. Если отклоняем, то предполагаем, что между одеялами есть разница, и следующим шагом должно быть выяснение, какое из одеял лучше.

Если выполняется предположение о нормальности распределения времени восстановления температуры, то лучше использовать Стьюдента, так как он использует больше информации о распределении и более точен.
